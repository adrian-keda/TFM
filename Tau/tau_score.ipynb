```python
import pandas as pd
import numpy as np
```

```python
expression = pd.read_csv('/home/amaqueda/TFM/ML_project/Tissue_Specificity/GTEx_Analysis_2017-06-05_v8_RNASeQCv1.1.9_gene_median_tpm.tsv', sep = '\t', header = 0)
print(expression.shape)
expression.head()
```

```python
expression['max'] = expression[expression.columns[2:]].max(axis = 1)
expression
```

```python
for column in expression.columns[2:-1]:
    expression[column] = expression[column]/expression['max']
expression
```

```python
expression['Tau_score'] = expression[expression.columns[2:-1]].sum(axis = 1) / 53
expression
```

```python
not_null = expression.loc[~expression[expression.columns[2:56]].isnull().all(axis=1)]
not_null
```

```python
not_null.rename(columns={'Description':'Symbol'})[['Name','Symbol','Tau_score']].to_csv('/home/amaqueda/TFM/ML_project/Tissue_Specificity/GTEx_Tau_Score_no_nulls.tsv', sep = '\t', index = False)
```
