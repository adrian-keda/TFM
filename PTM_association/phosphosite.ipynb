```python
import pandas as pd
import numpy as np
```

### MERGE INPUTS

```python
ace = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/input/PPS/Ace.tsv', sep = '\t', header = 0)
ace = ace[ace['ORGANISM'] == 'human']
ace['Residue'] = ace.MOD_RSD.str.split('-',expand = True)[0].str[0]
ace['Position'] = ace.MOD_RSD.str.split('-',expand = True)[0].str[1:]
ace['Type'] = 'Acetylation'
ace
```

```python
meth = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/input/PPS/Meth.tsv', sep = '\t', header = 0)
meth = meth[meth['ORGANISM'] == 'human']
meth['Residue'] = meth.MOD_RSD.str.split('-',expand = True)[0].str[0]
meth['Position'] = meth.MOD_RSD.str.split('-',expand = True)[0].str[1:]
meth['Type'] = 'Methylation'
meth
```

```python
phos = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/input/PPS/Phos.tsv', sep = '\t', header = 0)
phos = phos[phos['ORGANISM'] == 'human']
phos['Residue'] = phos.MOD_RSD.str.split('-',expand = True)[0].str[0]
phos['Position'] = phos.MOD_RSD.str.split('-',expand = True)[0].str[1:]
phos['Type'] = 'Phosphorylation'
phos
```

```python
ubi = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/input/PPS/Ubi.tsv', sep = '\t', header = 0)
ubi = ubi[ubi['ORGANISM'] == 'human']
ubi['Residue'] = ubi.MOD_RSD.str.split('-',expand = True)[0].str[0]
ubi['Position'] = ubi.MOD_RSD.str.split('-',expand = True)[0].str[1:]
ubi['Type'] = 'Ubiquitination'
ubi
```

```python
merge = pd.concat([ace,meth,phos,ubi], axis = 0)
merge
```

```python
merge[['GENE','ACC_ID','SITE_+/-7_AA','Residue','Position','Type','HU_CHR_LOC']].drop_duplicates()
```

```python
merge.rename(columns = {'HU_CHR_LOC':'Location'}, inplace = True)
```

```python
merge[['GENE','ACC_ID','SITE_+/-7_AA','Residue','Position','Type', 'Location']].drop_duplicates().to_csv('/home/amaqueda/TFM/ML_project/PTM_sites/output/PPS/PPS_clean.tsv', sep = '\t', index = False)
```

### MERGE UPDATED GENE SYMBOL

```python
pps = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/output/PPS/PPS_clean.tsv', sep = '\t', header = 0)
symbols = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/input/PPS/hgnc-symbol-check.csv', sep = ',', header = 0)
```

```python
symbols.rename(columns = {'Input':'GENE','Approved symbol':'Approved_symbol'}, inplace = True)
symbols.head()
```

```python
pps_clean = pps.merge(symbols[['GENE','Location','Approved_symbol']], how = 'inner', on = ['GENE', 'Location'])
pps_clean
```

```python
pps_clean[pps_clean['GENE'] != pps_clean['Approved_symbol']].to_csv('patata.tsv', sep = '\t', index = False)
```
