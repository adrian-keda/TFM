```python
print(ace.shape)
print(meth.shape)
print(phospho.shape)
print(ubi.shape)
```

```python
merge = pd.concat([ace,meth,phospho,ubi], axis = 0)
print(merge.shape)
merge
```

```python
merge.rename(columns = {0:'Uniprot_ID',1:'Uniprot_Accession',2:'AA_POS',3:'MOD_TYPE',4:'PubMed_ID',5:'Sequence'}, inplace = True)
merge.columns
```

```python
merge = merge[~merge['Sequence'].isnull()]
merge.shape
```

```python
merge['Residue'] = [i[10] for i in merge['Sequence'].tolist()]
merge
```

```python
merge[['Uniprot_ID', 'Uniprot_Accession', 'AA_POS', 'MOD_TYPE', 'Sequence','Residue']].to_csv('/home/amaqueda/TFM/ML_project/PTM_sites/output/dbPTM/all.tsv', sep = '\t', index = False)
```
