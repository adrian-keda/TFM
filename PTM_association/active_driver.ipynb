### LETS CLEAN ActiveDriver

```python
empty_dict = {'gene' : [],
              'position' : [],
              'residue' : [],
              'type': [],
              'kinase' : [],
              'refseq' : [],
              'uniprot' : []
              }


for pos,row in addb.iterrows():
    types = row['type'].split(',')
    for i in range(len(types)):
        if types[i] in ['phosphorylation','ubiquitination','methylation','acetylation']:
            empty_dict['gene'].append(row['gene'])
            empty_dict['position'].append(row['position'])
            empty_dict['residue'].append(row['residue'])
            empty_dict['type'].append(types[i])
            empty_dict['kinase'].append(row['kinase'])
            empty_dict['refseq'].append(row['refseq'])
            empty_dict['uniprot'].append(row['uniprot'])
```

```python
addb_clean = pd.DataFrame(data = empty_dict)
addb_clean
# This DataFrame has some sites with pmid = NaN
```

```python
addb_clean.to_csv('/home/amaqueda/TFM/ML_project/PTM_sites/output/ActiveDriver/addb.clean.tsv', sep = '\t', index = False)
```
