```python
import pandas as pd
import numpy as np
```

```python
ids = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/no_protein_level.tsv', sep = '\t', header = 0)
ids.rename(columns = {'Entry' : 'Accession','Entry Name' : 'Entry_Name','Gene Names (primary)' : 'Gene'}, inplace = True)
ids
```

```python
qPTM = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/output/qPTM/qPTM.clean.tsv', sep  = '\t')
qPTM_clean = qPTM.drop_duplicates(subset = ['Gene name','Position','PTM','UniProt accession'], keep = 'first')
print(qPTM_clean.shape)
qPTM_clean.rename(columns = {'Gene name':'Gene','PTM':'Type','UniProt accession':'Accession','residue':'Residue'}, inplace = True)
qPTM_merge = qPTM_clean.merge(ids, on = 'Accession')
qPTM_merge['Type'] = [i.lower() for i in qPTM_merge['Type']]
qPTM_merge
```

```python
qPTM_merge['Type'].value_counts()
```

```python
addb = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/output/ActiveDriver/addb.clean.tsv', sep = '\t')
addb.rename(columns = {'gene':'Gene','position':'Position','residue':'Residue','type':'Type','uniprot':'Accession'},inplace = True)
print(addb.shape)
addb_merge = addb.merge(ids, on = 'Accession')
addb_merge
```

```python
addb_merge['Type'].value_counts()
```

```python
dbPTM = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/output/dbPTM/all.tsv', sep = '\t')
dbPTM.rename(columns = {'Uniprot_ID':'Entry_Name','Uniprot_Accession':'Accession','AA_POS':'Position','MOD_TYPE':'Type'},inplace = True)
print(dbPTM.shape)
dbPTM_merge = dbPTM.merge(ids, on = 'Accession')
dbPTM_merge['Type'] = [i.lower() for i in dbPTM_merge['Type']]
dbPTM_merge.rename(columns = {'Gene':'Gene_y'}, inplace = True)
dbPTM_merge
```

```python
dbPTM_merge['Type'].value_counts()
```

```python
pps = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/output/PPS/PPS_clean.tsv', sep = '\t')
pps.rename(columns = {'GENE':'Gene','ACC_ID':'Accession'}, inplace = True)
print(pps.shape)
pps_merge = pps.merge(ids, on = 'Accession')
pps_merge['Type'] = [i.lower() for i in pps_merge['Type']]
pps_merge
```

```python
pps_merge['Type'].value_counts()
```

```python
uniprot = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/output/UniProt/UniProt.clean.tsv', sep = '\t')
uniprot.rename(columns = {'Entry' : 'Accession', 'Gene':'Gene_y'}, inplace = True)
uniprot
```
