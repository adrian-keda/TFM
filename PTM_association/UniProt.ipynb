```python
import pandas as pd
import numpy as np
```

```python
uniprot = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/input/UniProt/with_score.tsv', sep = '\t', header = 0)
print(uniprot.shape)
uniprot.head()
```

```python
uniprot_mod_res = uniprot[~uniprot['Modified residue'].isnull()]
print(uniprot_mod_res.shape)
uniprot_mod_res.head()
```

```python
# Maybe keep only 4 and 5 quality?
uniprot_mod_res['Annotation'].value_counts()
```

```python
empty_dict = {'Entry' : [],
              'Entry_Name' : [],
              'Gene' : [],
              'Position' : [],
              'Type' : [],
              'Sequence' : [],
              }

n_pass = 0

for a, row in uniprot_mod_res.iterrows():
    for i in row['Modified residue'].split('MOD_RES ')[1:]:

        try:
            res = i.split(';')
            position = int(res[0])
            modification = res[1]
            if not 'Microbial infection' in modification:
                if 'phospho' in modification or 'Phospho' in modification:
                    mod_type = 'phosphorylation'
                elif 'acetyl' in modification:
                    mod_type = 'acetylation'
                elif 'methyl' in modification:
                    mod_type = 'methylation'
                elif 'ubi' in modification:
                    print('UBIQUITILAION ???? -->', modification)
            
            empty_dict['Entry'].append(row['Entry'])
            empty_dict['Entry_Name'].append(row['Entry Name'])
            empty_dict['Gene'].append(row['Gene Names (primary)'])
            empty_dict['Position'].append(position)
            empty_dict['Type'].append(mod_type)
            empty_dict['Sequence'].append(row['Sequence'])


        except:
            print(i)
            n_pass += 1


print(n_pass)
# END
```

```python
print(len(empty_dict['Entry']),len(empty_dict['Entry_Name']),len(empty_dict['Gene']),len(empty_dict['Position']),len(empty_dict['Type']),len(empty_dict['Sequence']))
```

```python
ptm_uniprot = pd.DataFrame(data = empty_dict)
print(ptm_uniprot.shape)
ptm_uniprot.head()
```

```python
residues = []
for a, row in ptm_uniprot.iterrows():
    residues.append(row['Sequence'][row['Position'] - 1])
len(residues)
```

```python
ptm_uniprot['Residue'] = residues
ptm_uniprot.head()
```

```python
ptm_uniprot[['Entry','Entry_Name','Gene','Position','Residue','Type']].to_csv('/home/amaqueda/TFM/ML_project/PTM_sites/output/UniProt/UniProt.clean.tsv', sep = '\t', index = False)
```
