```python
import pandas as pd
import numpy as np
```

```python
qptm = pd.read_csv('/home/amaqueda/TFM/ML_project/PTM_sites/input/qPTM/qPTM_human_sites.tsv', sep = '\t', header = 0)
qptm.shape
```

```python
# Keep good quality sites (stars > 4)
stars = qptm[qptm.Reliability >= 4]
print(stars.shape)
stars.head()
```

```python
# As qPTM has proteomics data, drop duplicates to keep only the site
sites = stars[['Gene name','Position','PTM','Sequence window','UniProt accession']].drop_duplicates()
sites.head()
```

```python
sites.shape
```

```python
#All sites have +- 7 region
all([len(i) == 15 for i in sites['Sequence window']])
```

```python
sites['residue'] = [seq[7] for seq in sites['Sequence window'].tolist()]
print(sites.shape)
sites.head()
```

```python
sites.to_csv('/home/amaqueda/TFM/ML_project/PTM_sites/output/qPTM/qPTM.clean.tsv', sep = '\t', index = False)
```
