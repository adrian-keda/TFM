## SOME BASIC TESTS

```python
import pandas as pd

df = pd.read_csv('/home/amaqueda/TFM/ML_project/ML_part/input/PANCANCER_lof.fisher.permutation_input.tsv', sep = '\t', header = 0)
df['CPG'] = df['CPG'].fillna('No')
df.info()
```

```python
from scipy.stats import mannwhitneyu

for col in [s for s in df.columns if 'germline' in s]:
    print(col)
    print(mannwhitneyu(x = df[(df['CPG'] == 'Yes') & (~df[col].isnull())][col],
                       y = df[(df['CPG'] == 'No') & (~df[col].isnull())][col]), '\n')
```

```python
from scipy.stats import mannwhitneyu

for col in [s for s in df.columns if 'gnomAD' in s]:
    print(col)
    print(mannwhitneyu(x = df[(df['CPG'] == 'Yes') & (~df[col].isnull())][col],
                       y = df[(df['CPG'] == 'No') & (~df[col].isnull())][col]), '\n')
```

```python
# """
# Features not related to variants:
# - Tau scores
# - Network Distance
# - Prob of being Loss of Function intolerant
# """
from scipy.stats import mannwhitneyu

mannwhitneyu(x = df[(df['CPG'] == 'Yes') & (~df['tau_tissue_specificity'].isnull())]['tau_tissue_specificity'],
             y = df[(df['CPG'] == 'No') & (~df['tau_tissue_specificity'].isnull())]['tau_tissue_specificity'])
```

```python
from scipy.stats import mannwhitneyu

mannwhitneyu(x = df[(df['CPG'] == 'Yes') & (~df['oe_lof_upper'].isnull())]['oe_lof_upper'],
             y = df[(df['CPG'] == 'No') & (~df['oe_lof_upper'].isnull())]['oe_lof_upper'])
```

```python
from scipy.stats import mannwhitneyu

for col in [s for s in df.columns if ('amplification' in s)|('deletion' in s)]:
    print(col)
    print(mannwhitneyu(x = df[(df['CPG'] == 'Yes') & (~df[col].isnull())][col],
                       y = df[(df['CPG'] == 'No') & (~df[col].isnull())][col]), '\n')
```

```python
from scipy.stats import mannwhitneyu

for col in [s for s in df.columns if ('amplification' in s)|('deletion' in s)]:
    print(col,'--> CPGs:', df[(df['CPG'] == 'Yes') & (~df[col].isnull())][col].mean(),
          'random:', df[(df['CPG'] == 'No') & (~df[col].isnull())][col].mean(), '\n')
```

```python
from scipy.stats import mannwhitneyu

for col in [s for s in df.columns if ('loh' in s)|('value_EUR' in s)|('estimate' in s)]:
    print(col)
    print(mannwhitneyu(x = df[(df['CPG'] == 'Yes') & (~df[col].isnull())][col],
                       y = df[(df['CPG'] == 'No') & (~df[col].isnull())][col]), '\n')
```

```python
from scipy.stats import mannwhitneyu

for col in ['num.somatic_mutation', 'num.PTV_mutations', 'PTV_frequency']:
    print(col)
    print(mannwhitneyu(x = df[(df['CPG'] == 'Yes') & (~df[col].isnull())][col],
                       y = df[(df['CPG'] == 'No') & (~df[col].isnull())][col]), '\n')
```

## PLOTS

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df.columns
```

```python
plof = df[['Gene', 'oe_lof', 'CPG']]
plof.dropna(inplace = True)
plof_sort = plof.sort_values(by = 'oe_lof')
plof_sort.info()

cpg = plof_sort[plof_sort['CPG'] == 'Yes']
no_cpg = plof_sort[plof_sort['CPG'] == 'No'].sample(cpg.shape[0])
test = pd.concat([cpg, no_cpg]).sort_values('oe_lof')
test = test[test['oe_lof'] != 0]
test
```

```python
plt.figure(figsize = (8, 6))
a = plt.bar(x = np.linspace(1, 50, test.shape[0]), height = test['oe_lof'].tolist(),
        color = ['#f37735' if i == 'Yes' else '#00aedb' for i in test['CPG']],
        width = 0.19)
plt.title('Known CPG vs other genes', fontweight = 'bold', fontsize = 15)
plt.ylabel('Oe_lof', fontsize = 12.5)
plt.ylim(0,max(test['oe_lof']))
```

```python
avg = df[['Gene', 'avg_ALL', 'CPG']]
avg.dropna(inplace = True)
avg_sort = avg.sort_values(by = 'avg_ALL')
avg_sort.info()

cpg = avg_sort[avg_sort['CPG'] == 'Yes']
no_cpg = avg_sort[avg_sort['CPG'] == 'No'].sample(cpg.shape[0])
test = pd.concat([cpg, no_cpg]).sort_values('avg_ALL')
test = test[test['avg_ALL'] != 0]
test
```

```python
plt.figure(figsize = (8, 6))
plt.bar(x = np.linspace(1, 50, test.shape[0]), height = test['avg_ALL'].tolist(),
        color = ['#f37735' if i == 'Yes' else '#00aedb' for i in test['CPG']],
        width = 0.19)
plt.title('Known CPG vs other genes', fontweight = 'bold', fontsize = 15)
plt.ylabel('avg_ALL', fontsize = 12.5)
plt.ylim(0,max(test['avg_ALL']))
```
